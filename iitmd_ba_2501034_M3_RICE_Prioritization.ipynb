```python
import pandas as pd

features = pd.read_csv("data/processed/M2_feature_sentiment.csv")
features
```

```python
feature_list = [
    "Offline Mode",
    "Performance Bug Fixes",
    "Workout Tracking Enhancements",
    "UI / UX Improvements",
    "Wearable Sync Integration",
    "Subscription Pricing Optimization",
    "Customer Support Improvements",
    "Progress Tracking Enhancements",
    "Dark Mode",
    "Social / Community Features"
]

rice_df = pd.DataFrame({"Feature": feature_list})
rice_df
```

```python
reach_map = {
    "Offline Mode": 9000,
    "Performance Bug Fixes": 8000,
    "Workout Tracking Enhancements": 6000,
    "UI / UX Improvements": 5000,
    "Wearable Sync Integration": 4500,
    "Subscription Pricing Optimization": 4000,
    "Customer Support Improvements": 3500,
    "Progress Tracking Enhancements": 3000,
    "Dark Mode": 2000,
    "Social / Community Features": 2500
}

rice_df["Reach"] = rice_df["Feature"].map(reach_map)

impact_map = {
    "Offline Mode": 3,
    "Performance Bug Fixes": 3,
    "Workout Tracking Enhancements": 2,
    "UI / UX Improvements": 2,
    "Wearable Sync Integration": 2,
    "Subscription Pricing Optimization": 1,
    "Customer Support Improvements": 1,
    "Progress Tracking Enhancements": 1,
    "Dark Mode": 0.5,
    "Social / Community Features": 0.5
}

rice_df["Impact"] = rice_df["Feature"].map(impact_map)

confidence_map = {
    "Offline Mode": 1.0,
    "Performance Bug Fixes": 1.0,
    "Workout Tracking Enhancements": 0.9,
    "UI / UX Improvements": 0.8,
    "Wearable Sync Integration": 0.8,
    "Subscription Pricing Optimization": 0.7,
    "Customer Support Improvements": 0.7,
    "Progress Tracking Enhancements": 0.7,
    "Dark Mode": 0.6,
    "Social / Community Features": 0.5
}

rice_df["Confidence"] = rice_df["Feature"].map(confidence_map)

effort_map = {
    "Offline Mode": 4,
    "Performance Bug Fixes": 2,
    "Workout Tracking Enhancements": 2,
    "UI / UX Improvements": 1,
    "Wearable Sync Integration": 4,
    "Subscription Pricing Optimization": 1,
    "Customer Support Improvements": 1,
    "Progress Tracking Enhancements": 2,
    "Dark Mode": 1,
    "Social / Community Features": 3
}

rice_df["Effort"] = rice_df["Feature"].map(effort_map)
rice_df["RICE_Score"] = (
    rice_df["Reach"] *
    rice_df["Impact"] *
    rice_df["Confidence"] /
    rice_df["Effort"]
)

rice_df = rice_df.sort_values("RICE_Score", ascending=False)
rice_df

```

```python
rice_df["Effort_Double"] = rice_df["Effort"] * 2
rice_df["Effort_Double"]
print(rice_df["Effort_Double"])

```

```python
rice_df.to_csv(
    "data/processed/M3_RICE_Prioritization.csv",
    index=False
)

```
